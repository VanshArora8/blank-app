```python
# Calculating Q1 and Q3
Q1=df['Rotational speed [rpm]'].quantile(0.25)
Q3=df['Rotational speed [rpm]'].quantile(0.75)
# Finding IQR
IQR=Q3-Q1
# Calculating Outliers
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
new_df = df[(df['Rotational speed [rpm]'] >= ll) & (df['Rotational speed [rpm]'] <= up)]
new_df.reset_index(inplace=True)
new_df.drop(columns=['index'],inplace=True)
new_df
```

```python
Q1=new_df['Rotational speed [rpm]'].quantile(0.25)
Q3=new_df['Rotational speed [rpm]'].quantile(0.75)
IQR=Q3-Q1
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
new_df1 = new_df[(new_df['Rotational speed [rpm]'] >= ll) & (new_df['Rotational speed [rpm]'] <= up)]
new_df1.reset_index(inplace=True)
new_df1.drop(columns=['index'],inplace=True)
new_df1
```

```python
Q1=new_df1['Rotational speed [rpm]'].quantile(0.25)
Q3=new_df1['Rotational speed [rpm]'].quantile(0.75)
IQR=Q3-Q1
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
new_df2 = new_df1[(new_df1['Rotational speed [rpm]'] >= ll) & (new_df1['Rotational speed [rpm]'] <= up)]
new_df2.reset_index(inplace=True)
new_df2.drop(columns=['index'],inplace=True)
new_df2
```

```python
Q1=new_df2['Rotational speed [rpm]'].quantile(0.25)
Q3=new_df2['Rotational speed [rpm]'].quantile(0.75)
IQR=Q3-Q1
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
new_df3 = new_df2[(new_df2['Rotational speed [rpm]'] >= ll) & (new_df2['Rotational speed [rpm]'] <= up)]
new_df3.reset_index(inplace=True)
new_df3.drop(columns=['index'],inplace=True)
new_df3
```

```python
Q1=new_df3['Torque [Nm]'].quantile(0.25)
Q3=new_df3['Torque [Nm]'].quantile(0.75)
IQR=Q3-Q1
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
new_df4 = new_df3[(new_df3['Torque [Nm]'] >= ll) & (new_df3['Torque [Nm]'] <= up)]
new_df4.reset_index(inplace=True)
new_df4.drop(columns=['index'],inplace=True)
new_df4
```

```python
# Calculating Q1 and Q3
Q1=new_df4['Torque [Nm]'].quantile(0.25)
Q3=new_df4['Torque [Nm]'].quantile(0.75)
# Finding IQR
IQR=Q3-Q1
# Calculating Outliers
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
#Removing the Outliers
new_df5 = new_df4[(new_df4['Torque [Nm]'] >= ll) & (new_df4['Torque [Nm]'] <= up)]
# Reseting the index
new_df5.reset_index(inplace=True)
new_df5.drop(columns=['index'],inplace=True)
new_df5
```

```python
# Calculating Q1 and Q3
Q1=new_df5['Rotational speed [rpm]'].quantile(0.25)
Q3=new_df5['Rotational speed [rpm]'].quantile(0.75)
# Finding IQR
IQR=Q3-Q1
# Calculating Outliers
ll= Q1-1.5*IQR
up= Q3+1.5*IQR
#Removing the Outliers
new_df6 = new_df5[(new_df5['Rotational speed [rpm]'] >= ll) & (new_df5['Rotational speed [rpm]'] <= up)]
#Reseting the index
new_df6.reset_index(inplace=True)
new_df6.drop(columns=['index'],inplace=True)
new_df6
```

```python
plt.figure(figsize=(25, 15))

# Air temperature boxplot
plt.subplot(231)
sns.boxplot(x='Air temperature [K]', data=new_df6, color='cyan')
plt.title('Figure 3.1: Air temperature [K]')

# Process temperature Boxplot
plt.subplot(232)
sns.boxplot(x='Process temperature [K]', data=new_df6, color='y')
plt.title('Figure 3.2: Process temperature [K]')

# Rotational speed boxplot
plt.subplot(233)
sns.boxplot(x='Rotational speed [rpm]', data=new_df6, color='g')
plt.title('Figure 3.3: Rotational speed [rpm]')

# Torque boxplot
plt.subplot(234)
sns.boxplot(x='Torque [Nm]', data=new_df6, color='r')
plt.title('Figure 3.4: Torque [Nm]')

# Tool wear boxplot
plt.subplot(235)
sns.boxplot(x='Tool wear [min]', data=new_df6, color='pink')
plt.title('Figure 3.5: Tool wear [min]')

plt.tight_layout()
plt.show()
```

```python
# independent & dependent features
x=new_df6.iloc[:,0:5]
y=new_df6.iloc[:,-1]
```

```python
# training set 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=42)
```

```python
#model building 
from sklearn.linear_model import LogisticRegression
```

```python
# Using Logistic Regression
classifier = LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print("The Accuracy is %.2f" % score)
s=classification_report(y_pred,y_test)
print("Classification Report")
print(s)
```
